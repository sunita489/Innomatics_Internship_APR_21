### Mean
Mean is described as the total sum of the numbers in a list divided by the length of the numbers in the list.

Mathematically, mean is calculated by below formula:

$ \mu = $ ($\sum_{i=0}^{n} a_{j}$)  / n

In python, **Statistics** library is used to carry out mean of given data



```python
#Calculating mean by mathematical formula
print(sum(df.loc[:,"Mthly_HH_Income"])/len(df.loc[:,"Mthly_HH_Income"]))
print(sum(df.loc[:,"Mthly_HH_Expense"])/len(df.loc[:,"Mthly_HH_Expense"]))
```

```python
#Calculating mean using statiscs library
import statistics
print(statistics.mean(df.loc[:,"Mthly_HH_Income"]))
print(statistics.mean(df.loc[:,"Mthly_HH_Expense"]))
```

### Median

Mean is described as middle value of the given list when arranges in an order.

First sort the values in increasing order ,then divide the length of sorted list by 2.

Mathematically , below formula is used :

Med(x) = X  [$ \frac{n}{2} \\ $]&nbsp;&nbsp;&nbsp;&nbsp;  if n is even.

Med(x) = (X [$ \frac{n-1}{2} \\ $] + X [$ \frac{n+1}{2} \\ $]) / 2 &nbsp;&nbsp;&nbsp;&nbsp;if n is even.

In python, **Statistics** library is used to determine median of given data.


```python
#Arranging the "Mthly_HH_Income" in ascending order
df.sort_values(by=['Mthly_HH_Income'], inplace=True)

#Accessing the middle value from sorted list.
print(df.iloc[len(df.loc[:,"Mthly_HH_Income"])//2]['Mthly_HH_Income'])

#Arranging the "Mthly_HH_Expense" in ascending order
df.sort_values(by=['Mthly_HH_Expense'], inplace=True)

#Accessing the middle value from sorted list.
print(df.iloc[len(df.loc[:,"Mthly_HH_Expense"])//2]['Mthly_HH_Income'])
```

```python
#Calculating median using statiscs library
import statistics
print(statistics.median(df.loc[:,"Mthly_HH_Income"]))
print(statistics.median(df.loc[:,"Mthly_HH_Expense"]))
```

### Mode

Mode is the value with highest frequency when arranged is ascending order.

When there are two modes in a data set, then the set is called bimodal.

When there are three modes in a data set, then the set is called trimodal

When there are four or more modes in a data set, then the set is called multimodal.

```python
#Calculating mode using statiscs library
import statistics
print(statistics.mode(df.loc[:,"No_of_Fly_Members"]))
print(statistics.mode(df.loc[:,"Highest_Qualified_Member"]))
print(statistics.mode(df.loc[:,"No_of_Earning_Members"]))
```

### Measure of spread / Dispersion

It is used to determine how data is spread obout the mean.

### Variance ($ \sigma^2 $)

Variance is the mean of square of difference between mean and each datapoints in given data.
Mathematically , it is given as :

$ \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n} $

where ,

**xi** = datapoint in data

$ \mu $ = mean

**n** = total number of datapoints

More the variance more is data spread and viceversa with respect to mean. 

```python
#Calculating variance by mathematical formula
lst = list(df.loc[:,"Mthly_HH_Income"])
lst1 = list(df.loc[:,"Mthly_HH_Expense"])
total = 0
total1 = 0

#Mean of "Mthly_HH_Income" and "Mthly_HH_Expense"
x = sum(df.loc[:,"Mthly_HH_Income"])/len(df.loc[:,"Mthly_HH_Income"])
x1 = sum(df.loc[:,"Mthly_HH_Expense"])/len(df.loc[:,"Mthly_HH_Expense"])

for i in range(len(lst)):
    total = total + abs(x-lst[i])**2
    total1 = total1 + abs(x1-lst1[i])**2
    
print(total/len(lst))
print(total1/len(lst1))

```

```python
#Calculating mode using statiscs library
import statistics
print(statistics.pvariance(df.loc[:,"Mthly_HH_Income"]))
print(statistics.pvariance(df.loc[:,"Mthly_HH_Expense"]))
```

### Standard Deviation($ \sigma $)

Standard Deviation is simply square root of Variance.

It also shows how much data is spread with respect to mean.

Smaller the SD less the points are spread and viceversa. 

Mathematically it is calculated using:

$ \sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}} $

where ,

**xi** = datapoint in data

$ \mu $ = mean

**n** = total number of datapoints


```python
#Calculating standard devaition by mathematical formula
lst = list(df.loc[:,"Mthly_HH_Income"])
lst1 = list(df.loc[:,"Mthly_HH_Expense"])
total = 0
total1 = 0

#Mean of "Mthly_HH_Income" and "Mthly_HH_Expense"
x = sum(df.loc[:,"Mthly_HH_Income"])/len(df.loc[:,"Mthly_HH_Income"])
x1 = sum(df.loc[:,"Mthly_HH_Expense"])/len(df.loc[:,"Mthly_HH_Expense"])

for i in range(len(lst)):
    total = total + abs(x-lst[i])**2
    total1 = total1 + abs(x1-lst1[i])**2
    
print((total/len(lst))**0.5)
print((total1/len(lst1))**0.5)
```

```python
#Calculating mode using statiscs library
import statistics
print(statistics.pstdev(df.loc[:,"Mthly_HH_Income"]))
print(statistics.pstdev(df.loc[:,"Mthly_HH_Expense"]))
```

### Correlation

Correlation is a statistical technique used to determine relationship between two variables.

It can be measured using various coefficients.

1.Pearson's cofficient.

2.Spearman's rank correlation coefficient.

Most commonly used coefficient is Pearson's cofficient also called as sampling correlation coefficient.

Mathematically , Pearson's cofficient is calculated by using below formula:

r = $ \frac{\sum (x - \bar{x})(y - \bar{y})}{\sqrt{\sum (x - \bar{x})^2 \sum (y - \bar{y})^2}}   $

where , 

**x** = datapoint of first variables/column.

**y** = datapoint of second variables/column.

$\bar{x}$ = mean of first variables/column.

$\bar{y}$ = mean of second variables/column.

Pearson's Coefficient can be calculated by using pandas as well as numpy.

Pandas - df.corr()

Numpy - df.corrcef(x,y)

```python
lst1 = list(df.loc[:,"Mthly_HH_Income"])
lst2 = list(df.loc[:,"Mthly_HH_Expense"])
t1=0
t2=0
t3=0

x = sum(df.loc[:,"Mthly_HH_Income"])/len(df.loc[:,"Mthly_HH_Income"])
y = sum(df.loc[:,"Mthly_HH_Expense"])/len(df.loc[:,"Mthly_HH_Expense"])

for i in range(len(df.loc[:,"Mthly_HH_Expense"])):
    t1 = t1 + ((x - lst1[i])*(y - lst2[i]))
    t2 = t2 + (x - lst1[i])**2
    t3 = t3 + (y - lst2[i])**2
t4 = (t2*t3)**0.5
print(round(t1/t4,5))
    
```

### Normal Distribution.

Normal distribution is a bell-shaped curve which describes how the variables are distributed.

Here data is centered about the mean, median and mode.

Properties of normal distribution.

1. Bell-shaped curve which has perfect bilateral symmetry.

2. Symmetrical in shape.

3. It follows 68 - 95 - 99.7% rule.


68 - 95 - 99.7% rule:

68% of data lies in $ \mu $ + $\sigma$ and $ \mu $ - $\sigma$.

95% of data lies in $ \mu $ + 2$\sigma$ and $ \mu $ - 2$\sigma$.

99.7% data lies in $ \mu $ + 3$\sigma$ and $ \mu $ - 3$\sigma$.

Generally normal distribution is shown by using:

f(x, $\mu$, $\sigma$) =$ \frac{1}{\sigma\sqrt{2\pi}}\exp\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{2}\right)$

where :

$\mu$ = mean

$\sigma$ = standard deviation.

and x, $\mu$, $\sigma$ are parameters.

Exactly half of the values are to the left of the curve and half to right of the curve.

Area under the curve is always 1.



```python
col1_mean = col1.mean()
col1_std = col1.std()
print(col1_mean)
print(col1_std)
```

```python
#Checking 65 - 95 - 99.7% rule
l_65 = col1_mean - col1_std
r_65 = col1_mean + col1_std

l_95 = col1_mean - 2*col1_std
r_95 = col1_mean + 2*col1_std

l_997 = col1_mean - 3*col1_std
r_997 = col1_mean + 3*col1_std

plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(col1)

plt.axvline(col1_mean, color='coral', label='Mean')

plt.axvline(r_65, color='yellow', label='Mean + 1SD')
plt.axvline(l_65, color='yellow', label='Mean - 1SD')
plt.axvline(r_95, color='green', label='Mean + 2SD')
plt.axvline(l_95, color='green', label='Mean - 2SD')
plt.axvline(r_997, color='blue', label='Mean + 3SD')
plt.axvline(l_997, color='blue', label='Mean - 3SD')
plt.legend();
```

### Skewness

Skewness is the measure of asymmetry of the probability distribution of random variable about its mean.

In short it tells whether distribution is symmetric or not?

**Distributions can be of different types :**

1. Positively skewed distribution(right)

2. Negatively skewed distribution(left)

3. Undefined

If skewness is zero then distribution is perfectly symmetrical.

**If skewness is :**

1. less than -1 or greater than +1 then distribution is highly skewed.

2. between -1 to -0.5 or +1 to +0.5 then distribution is moderately skewed.

3. between -0.5 and 0.5 then distribution is approximately symmetric.

```python
df["Mthly_HH_Income"].skew()
```

```python
df["Mthly_HH_Expense"].skew()
```

```python
df["No_of_Fly_Members"].skew()
```

```python
df["Emi_or_Rent_Amt"].skew()
```

### Effect on Mean, Median and Mode due to Skewness

In a perfectly symmetrical distribution, the mean and the median are the same.

If mean is less than the median, and they are both less than the mode then distribution is skewed to left.

If mean is less than the median, and they are both less than the mode then distribution is skewed to right.

### Box Cox Transformation.

Box cox transformation helps to convert non-normal distribution into almost normal distribution.

In statistics, we generally assume that distributions should be in normal or Guassian form.Thus this is the effective way.



```python
pareto_col = np.random.pareto(a = 2, size = 1000)

sns.distplot(pareto_col)
```

```python
# Normality Test

stats.probplot(pareto_col, dist="norm", plot=plt)

plt.grid()
```

```python
# x_t = tranformed by box-cox, l = Lambda

x_t, l = stats.boxcox(pareto_col)

print(l)
```

```python
stats.probplot(x_t, dist="norm", plot=plt)

plt.grid()
```
