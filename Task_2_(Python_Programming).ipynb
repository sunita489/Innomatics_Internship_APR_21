```python
#Question 5
if __name__ == '__main__':
    N = int(input())
    lst1 = []
    for i in range(N):
        lst = list((input().split()))
        if lst[0] == 'insert':
            lst1.insert(int(lst[1]),int(lst[2]))
        if lst[0] == 'print':
            print(lst1)
        if lst[0] == 'remove':
            ele = int(lst[1])
            lst1.remove(ele)
        if lst[0] == 'append':
            lst1.append(int(lst[1]))
        if lst[0] == 'sort':
            lst1.sort()
        if lst[0] == 'reverse':
            lst1.reverse()
        if lst[0] == 'pop':
            lst1.pop()
```

```python
#Question 11
n = int(input())
s = set(map(int, input().split()))
su=int()
for _ in range(int(input())):
    l=list(input().split())
    if(l[0]=="pop"):
        s.pop()
    elif(l[0]=="remove"):
        s.remove(int(l[1]))
    elif(l[0]=="discard"):
        s.discard(int(l[1]))
for j in s:
    su=su+j
print(su)
```

```python
#Question 16
a = int(input())
A = set(map(int,input().split(' ')))
t = int(input())
for i in range(t):
    op = input().split()
    if op[0] == 'update':
        A.update(set(map(int,input().split(' '))))
    if op[0] == 'intersection_update':
        A.intersection_update(set(map(int,input().split(' '))))
    if op[0] == 'symmetric_difference_update':
        A.symmetric_difference_update(set(map(int,input().split(' '))))
    if op[0] == 'difference_update':
        A.difference_update(set(map(int,input().split(' '))))
print(sum(A))
```
