```python
#Question 4
import re
vowels = "[aeiou]{2,}"
consonants = "[^aeiou]" 
pattern = r"(?<={})({}){}".format(consonants, vowels, consonants)
matched_pattern = re.findall(pattern, input().replace(" ", ""), re.I)
print("\n".join(matched_pattern) if matched_pattern else -1)
```

```python
#Question 6
import re

def change(match):
    symb = match.group(0)
    
    if sym == "&&":
        return "and"
    elif sym == "||":
        return "or"
    
n = int(input().strip())
for _ in range(n):
    print(re.sub(r'(?<= )(&&|\|\|)(?= )', change, input()))
```

```python
#Question 10
import re

n = int(input().strip())
inside = False
for _ in range(n):
    line = input()
    
    for el in line.split(' '):
        if el == "{":
            inside = True
            continue
        elif el == "}":
            inside = False
            continue
        elif inside:
            found = re.search(r'\#[0-9a-fA-F]{3,6}', el)
            if found:
                print(found.group(0))
```

```python
#Question 12
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
  
 def handle_comment(self, data):
        if data.count('\n') > 0:
            print(">>> Multi-line Comment")
        else:
            print(">>> Single-line Comment")
        print(data)
 def handle_data(self, data):
        if len(data) > 1:
            print(">>> Data")
            print(data)
  
  
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Question 14
import re

if __name__ == "__main__":
    t = int(input().strip())
    
    for _ in range(t):
        uid = "".join(sorted(input()))
        if (len(uid) == 10 and
            re.match(r'', uid) and 
            re.search(r'[A-Z]{2}', uid) and
            re.search(r'\d\d\d', uid) and
            not re.search(r'[^a-zA-Z0-9]', uid) and
            not re.search(r'(.)\1', uid)):
            print("Valid")
        else:
            print("Invalid")
                
```

```python
#Question 15
import re
for i in range(int(input())):
    S = input().strip()
    pre_match = re.search(r'^[456]\d{3}(-?)\d{4}\1\d{4}\1\d{4}$',S)
    if pre_match:
        processed_string = "".join(pre_match.group(0).split('-'))
        final_match = re.search(r'(\d)\1{3,}',processed_string)
        if final_match:
            print('Invalid')
        else :
            print('Valid')
    else:
        print('Invalid')
```
