```python
#Question 7
import numpy
n = input().split()
list1 = []
list2 = []
for i in range(int(n[0])):
    a = map(int,input().split())
    list1.append(list(a))
for j in range(int(n[0])):
    b = map(int,input().split())
    list2.append(list(b))
a = numpy.array(list1)
b = numpy.array(list2)

print(a+b)
print(a-b)
print(a*b)
print(a//b)
print(a%b)
print(a**b)
```
